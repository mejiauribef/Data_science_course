```python
import numpy as np
import matplotlib.pyplot as plt

# Probability of winning El Baloto in Colombia
baloto_prob = 1 / 8145060

# Probability of winning a giveaway by Mr. Beast
mr_beast_prob = 5 / 15500000

# Generate data for the two graphs
x = ['El Baloto', 'Mr. Beast Giveaway']
y = [baloto_prob, mr_beast_prob]

# Set up the figure and subplot axes
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Generate the El Baloto graph
axs[0].bar(x[0], y[0], color='blue', alpha=0.8)
axs[0].set_ylabel('Probability')
axs[0].set_title('Probability of Winning El Baloto')
axs[0].set_yscale('log')
axs[0].set_ylim([1e-8, 1e0])

# Generate the Mr. Beast graph
axs[1].bar(x[1], y[1], color='green', alpha=0.8)
axs[1].set_ylabel('Probability')
axs[1].set_title('Probability of Winning a Mr. Beast Giveaway')
axs[1].set_yscale('log')
axs[1].set_ylim([1e-18, 1e-13])

# Add horizontal lines to compare the probabilities visually
axs[0].axhline(y=mr_beast_prob, color='gray', linestyle='--', alpha=0.8)
axs[1].axhline(y=baloto_prob, color='gray', linestyle='--', alpha=0.8)

# Add labels to the horizontal lines
axs[0].text(x[0], mr_beast_prob, 'Mr. Beast Giveaway\nProbability', ha='center', va='bottom', fontsize=10)
axs[1].text(x[1], baloto_prob, 'El Baloto\nProbability', ha='center', va='bottom', fontsize=10)

# Display the graphs side by side
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Probability of winning El Baloto in Colombia
baloto_prob = 1 / 8145060

# Probability of winning a giveaway by Mr. Beast
mr_beast_prob = 5 / 15500000

# Generate data for the graph
x = ['El Baloto', 'Mr. Beast Giveaway']
y = [baloto_prob, mr_beast_prob]

# Set up the figure and subplot axes
fig, ax = plt.subplots(figsize=(10,6))

# Generate the bar graph
ax.bar(x, y, color=['blue', 'green'], alpha=0.8)

# Add labels to the bars
ax.text(x[0], y[0]/2, '1 in 8 million', ha='center', va='center', fontsize=15, color='white')
ax.text(x[1], y[1]/2, '1 in 600 trillion', ha='center', va='center', fontsize=15, color='white')

# Add a title and labels to the axes
ax.set_title('Your Chances of Winning "El Baloto" vs. Winning a Giveaway by Mr. Beast', fontsize=20)
ax.set_ylabel('Probability', fontsize=15)

# Add a grid and remove the top and right spines
ax.grid(axis='y', alpha=0.5, linestyle='--')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add a legend
#ax.legend(['El Baloto', 'Mr. Beast Giveaway'], loc='upper right', fontsize=15)

# Display the graph
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

baloto_probs = np.array([7.643e-08, 3.822e-06, 1.529e-07, 7.643e-06, 3.822e-08, 1.912e-05, 7.643e-08, 3.822e-05, 6.143e-07])
mrbeast_probs = np.array([5/15.5e6]*9)

labels = ['5 + SB', '5', '4 + SB', '4', '3 + SB', '3', '2 + SB', '2', '1 + SB']

# Bar graph
plt.bar(labels, baloto_probs, color='blue', label='Baloto')
plt.bar(labels, mrbeast_probs, color='orange', label="Mr. Beast's Giveaway", alpha=0.5)
plt.xlabel('Prize Categories')
plt.ylabel('Probability')
plt.legend()
plt.title('Bar Graph: Baloto vs. Mr. Beast\'s Giveaway Probabilities')
plt.show()

# Pie chart
combined_probs = np.vstack((baloto_probs, mrbeast_probs)).T
colors = ['blue', 'orange']
explode = (0.1, 0)
fig, axes = plt.subplots(3, 3, figsize=(10, 10))
for i, ax in enumerate(axes.flatten()):
    ax.pie(combined_probs[i], labels=['Baloto', 'Mr. Beast'], autopct='%1.1f%%', colors=colors, explode=explode)
    ax.set_title(labels[i])
plt.suptitle('Pie Chart: Winning Probabilities Distribution', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Line graph
x = np.arange(len(labels))
plt.plot(x, baloto_probs, '-o', label='Baloto', color='blue')
plt.plot(x, mrbeast_probs, '-o', label="Mr. Beast's Giveaway", color='orange')
plt.xlabel('Prize Categories')
plt.ylabel('Probability')
plt.xticks(x, labels)
plt.legend()
plt.title('Line Graph: Cumulative Probabilities of Baloto and Mr. Beast\'s Giveaway')
plt.show()

```
