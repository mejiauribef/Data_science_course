Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Deficion f(x)

```python
def f(x):
    return x**2
x = np.linspace(-10,10,num=1000)
y = f(x)

plt.plot(x,y)
plt.grid()
```

Derivando f(x)

```python
def df(f):
    h=0.000001
    return (f(x+h)-f(x))/h

plt.plot(x,y, 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```

Probando con otras funciones

```python
def f(x):
    return np.sin(x)
y = f(x)

plt.plot(x,y, 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```
