```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA','JAPAN','MEXICO','COLOMBIA','GERMANY']
population = [1000,800,750,450,400]
```

```python
fig, ax = plt.subplots()
fig.set_facecolor('#F5F5F5')
ax.bar(country,population,width=0.5,color=['#FF9933','#009246','#FFC72C','#000000','#E60012'])
plt.xticks(np.arange(5),('India','Mexico','Colombia','Alemania','Japon'),rotation=45)
plt.title('Population per country 2009')
plt.xlabel('Countries')
plt.ylabel('Popuplation per millions')
plt.show()
```

```python
plt.barh(country,population)
plt.show()
```
