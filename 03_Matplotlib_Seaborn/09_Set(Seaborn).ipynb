```python
# Crear datos aleatorios
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

# Calcular valores de seno y coseno
Z1 = np.sin(np.sqrt(X**2 + Y**2))
Z2 = np.cos(np.sqrt(X**2 + Y**2))

# Crear figura y eje tridimensional
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Graficar superficies de seno y coseno
surf1 = ax.plot_surface(X, Y, Z1, cmap=cm.Purples, alpha=0.7, linewidth=0, antialiased=False)
surf2 = ax.plot_surface(X, Y, Z2, cmap=cm.Blues, alpha=0.7, linewidth=0, antialiased=False)

# Configurar colorbar
fig.colorbar(surf1, shrink=0.5, aspect=10)

# Agregar etiquetas y título
ax.set_xlabel('X', fontsize=14)
ax.set_ylabel('Y', fontsize=14)
ax.set_zlabel('Z', fontsize=14)
ax.set_title('Visualización en 3D de Seno y Coseno', fontsize=18)

# Configurar vista y rotación
ax.view_init(elev=30, azim=45)

# Mostrar la figura
plt.show()
```
